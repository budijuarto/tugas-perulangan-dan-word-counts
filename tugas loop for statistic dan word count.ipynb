```python
import random
nilaitugas = []
for l in range(250):
    nilaitugas.append(random.randrange(1,100,1))
print(nilaitugas)
print("jumlah nilai tugas adalah", len(nilaitugas))
```

```python
#bilangan ganjil
for i in nilaitugas:
    if i % 2 == 1:
        print(i)
       
```

```python

for a in nilaitugas:
    if a > 70:
        print(a)
```

```python
#mencari simpangan baku
#mencari total xi
x2=0
for x in nilaitugas:
    x = x2+ x
k = x**2
#mencari xi^2
for z in nilaitugas:
    y = 0
    z = y + (z**2)
#menghitung var dan stdev
var = ((len(nilaitugas)*k)-(z))/((len(nilaitugas))*((len(nilaitugas))-1))
stdev = var ** (1/2)
print('Standar deviasi yang diperoleh yaitu', stdev)
```

```python
#menghitung nilai maksimum
maxs = nilaitugas[0]
for subs1 in nilaitugas:
    if maxs < subs1:
        maxs = subs1
print(maxs)
#menghitung nilai minimum
mins = nilaitugas[0]
for subs2 in nilaitugas:
    if mins > subs2:
        mins = subs2
print (mins)
```

```python
# 5. Menghitung modus dari nilai 

#membuat dictionaries
book = {}
for frek in nilaitugas:
    if frek in book:
        book[frek] = book[frek] + 1
    else: 
        book[frek] = 1

#nyari nilai modus dgn cari paling tinggi value nya
high=0
mod = 0
for frek in book:
    if book[frek] > high:
        high = book[frek]
        mod = frek
        
print('modus yang diperoleh adalah', mod, 'yaitu', high)
```

```python
paragraf = "Barcelona berhasil merebut posisi puncak klasemen La Liga Spanyol seusai menang 4-2 atas Sevilla pada pertandingan pekan kesembilan di Stadion Camp Nou, Sabtu (20/10/2018) atau Minggu dini hari WIB. Barcelona membuka keunggulan pada menit ke-2 melalui gol yang dicetak oleh Philippe Coutinho. Lionel Messi menggandakan keunggulan Barcelona, 10 menit berselang.  Namun, nahas bagi Barcelona, karena Messi harus ditarik keluar pada menit ke-26 setelah mengalami cedera. Tanpa Messi, skor 2-0 bertahan hingga babak pertama berakhir."
pisah = paragraf.split()
wordcount={}
for n in pisah:
    if n not in wordcount:
        wordcount[n] = 1
    else:
        wordcount[n] += 1
#print (wordcount)
for k,v in wordcount.items():
    print(k,v)
```
